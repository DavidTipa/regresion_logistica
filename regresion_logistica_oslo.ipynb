```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.special as special
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('listings_50_outliers_oslo.csv')
df.head(20)
```

```python
#df = df.drop(columns=['Unnamed: 0'])
df
```

```python
df=df.fillna(method='bfill')
df=df.fillna(method='ffill')
```

```python
Var_Indep = df[['price', 'reviews_per_month', 'minimum_nights']]
Var_Dep = df['host_is_superhost']
```

```python
unic=df['price'].unique()
unic
```

```python
X=Var_Indep
y=Var_Dep
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=None)
```

```python
escalar=StandardScaler()
```

```python
X_train=escalar.fit_transform(X_train)
X_test=escalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
algoritmo=LogisticRegression()
```

```python
algoritmo.fit(X_train,y_train)
```

```python
y_pred=algoritmo.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
matriz=confusion_matrix(y_test,y_pred)
print('matriz de confusion')
print(matriz)
```

```python
from sklearn.metrics import precision_score
precision=precision_score(y_test,y_pred, average='binary', pos_label='f')
print('precision')
print(precision)
```

```python
from sklearn.metrics import accuracy_score
exactitud=accuracy_score(y_test,y_pred)
print('exactitud del modelo')
print(exactitud)
```
